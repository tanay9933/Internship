# Importing Packages

```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
import re

import warnings
warnings.filterwarnings ('ignore')
```

# Connecting to WEBDRIVER

```python
driver = webdriver.Chrome(r'C:\Users\rjhcl\Desktop\web driver\chromedriver.exe')
```

# Opening Wikipedia

```python
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
```

# Create Empty LIST

```python
Rank = []
Name = []
Artist = []
Upload_date = []
Views = []
```

A) Rank
B) Name
C) Artist
D) Upload date
E) Views

# SCRAPING the Data

```python
rank = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
for i in rank:
    Rank.append(i.text)
```

```python
name = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
for i in name:
    Name.append(i.text)
```

```python
artist = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
for i in artist:
    Artist.append(i.text)
```

```python
upload_date = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
for i in upload_date:
    Upload_date.append(i.text)
```

```python
views = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
for i in views:
    Views.append(i.text)
```

# Checking LENGTH of the LISTS

```python
print(len(Rank), len(Name), len(Artist), len(Upload_date), len(Views))
```

# Creating the DATAFRAME

```python
df=pd.DataFrame({'Index':Rank, 'Video Name':Name, 'Artist': Artist, 'Date of Upload':Upload_date, 'Total Views': Views})
                 
df
```
